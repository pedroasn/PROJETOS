```python
URLS = {
    "bolsa": URL_BOLSA,
    "cadunico": URL_CADUNICO,
    "pcd_bpc": URL_PCD_BPC,
    "terreiro": URL_TERREIRO,
    "idoso_bpc": URL_IDOSO_BPC,
    "indigenas": URL_INDIGENAS,
    "quilombolas": URL_QUILOMBOLAS,
    "situacao_rua": URL_SITUACAO_RUA,
}
```

```python
for nome_arquivo, url in URLS.items():

    print(f"\nBaixando {nome_arquivo}.csv...")

    try:
        response = requests.get(url, stream=True, timeout=600)
        response.raise_for_status()

        # Arquivo temporário
        tmp_file = f"/tmp/{nome_arquivo}.csv"

        with open(tmp_file, "wb") as f:
            for chunk in response.iter_content(chunk_size=1024 * 1024):
                if chunk:
                    f.write(chunk)

        destino = f"{files_path}/{nome_arquivo}.csv"

        # Remove o arquivo antigo (caso exista)
        try:
            mssparkutils.fs.rm(destino)
        except:
            pass

        # Copia para o Lakehouse
        mssparkutils.fs.cp(
            f"file:{tmp_file}",
            destino
        )

        print(f"✔ {nome_arquivo}.csv salvo com sucesso.")

    except Exception as e:
        print(f"❌ Erro ao baixar {nome_arquivo}: {e}")

print("\nProcesso concluído.")
```

```python
df = df.withColumnRenamed("Código", "Código_municipio") \
       .withColumnRenamed("Unidade Territorial", "Unidade_Territorial") \
       .withColumnRenamed("UF", "UF") \
       .withColumnRenamed("Referência", "Data") \
       .withColumnRenamed("Quantidade de famílias quilombolas com renda per capita mensal até meio salário-mínimo (Pobreza + Baixa renda) inscritas no Cadastro Único", "quilombolas_inscritos")


df = df.select("Código_municipio","Unidade_Territorial" ,"UF", "Data", "quilombolas_inscritos")

df = df.withColumn("quilombolas_inscritos", col("quilombolas_inscritos").cast("int"))

display(df)  
```

```python
# NOME DA TABELA DE DESTINO
tabela = 'tbQuilombolas'


# CAMINHO COMPLETO (LAKE + SCHEMA + TABELA)
path_tabela = f"{tables_path}/vul/{tabela}"


# SALVAR TABELA DELTA
df.write.format("delta") \
    .mode("overwrite") \
    .save(path_tabela)
```

## **DISQUE 100**  

```python
# Data atual
hoje = datetime.now()

ano = hoje.year
semestre = "primeiro-semestre" if hoje.month <= 6 else "segundo-semestre"

# Nome do arquivo
nome_arquivo = f"disque100-{semestre}-{ano}"

# URL
url = (
    f"https://www.gov.br/mdh/pt-br/acesso-a-informacao/dados-abertos/"
    f"disque100/{nome_arquivo}.csv"
)

print(nome_arquivo)
print(url)

# Download
print(f"\nVerificando disponibilidade de {nome_arquivo}.csv...")

try:
    # Fazemos uma requisição GET com stream=True
    response = requests.get(url, stream=True, timeout=600)
    
    # Se o status for 404 (Não Encontrado) ou 403 (Forbidden, comum quando o arquivo não existe em servidores gov)
    if response.status_code in [403, 404]:
        print(f"⚠️ O arquivo {nome_arquivo}.csv ainda não está disponível no servidor (Status: {response.status_code}). Seguindo em frente...")
    
    else:
        # Garante que qualquer outro erro de rede (ex: 500 erro interno) ainda seja tratado como falha real
        response.raise_for_status()

        print(f"📥 Arquivo encontrado! Baixando {nome_arquivo}.csv...")
        tmp_file = f"/tmp/{nome_arquivo}.csv"

        with open(tmp_file, "wb") as f:
            for chunk in response.iter_content(chunk_size=1024 * 1024):
                if chunk:
                    f.write(chunk)

        destino = f"{files_path}/{nome_arquivo}.csv"

        # Remove o arquivo antigo
        try:
            mssparkutils.fs.rm(destino)
        except:
            pass

        # Copia para o Lakehouse
        mssparkutils.fs.cp(
            f"file:{tmp_file}",
            destino
        )

        print(f"✔ {nome_arquivo}.csv salvo com sucesso.")

except Exception as e:
    # Este bloco agora só capturará problemas reais de rede, erros 500 ou falhas na cópia para o Lakehouse
    print(f"❌ Erro inesperado ao processar {nome_arquivo}: {e}")

print("\nProcesso concluído.")
```

```python
## PADRÃO DO ARQUIVO (Começa com disque100 e termina com .csv)
padrao_arquivo = "disque100*.csv"

## LEITURA
# O Spark vai varrer a pasta e ler todos os arquivos que batem com o padrão
df = spark.read.format("csv") \
    .option("header", "true") \
    .option("sep", ";") \
    .option("encoding", "utf-8") \
    .load(f'{files_path}/{padrao_arquivo}')

# Normaliza os nomes das colunas (substituindo espaços por _)
for col in df.columns:
    novo_nome = col.replace(" ", "_")
    df = df.withColumnRenamed(col, novo_nome)

# filtrando PE e Terreiro

df = df.filter(df["UF"] == "PE") \
       .filter(F.lower(F.col("Etnia_da_vítima")).contains("terreiro"))


# Remove duplicados
df = df.dropDuplicates(["hash"])      
        

# selecionando as colunas 

df = df.select("Data_de_cadastro","Município" ,"sl_quantidade_vitimas", "Etnia_da_vítima", "UF", "hash")


df = df.withColumn("temp_split", F.split(F.col("Município"), "\|"))


df = df \
    .withColumn("cod_municipio", F.trim(F.col("temp_split").getItem(0))) \
    .withColumn("municipio", F.trim(F.col("temp_split").getItem(1))) \
    .drop("temp_split")  # Remove a coluna temporária de array

df = df.select("hash", "Data_de_cadastro","municipio" ,"UF", "cod_municipio" ,"sl_quantidade_vitimas", "Etnia_da_vítima")



display(df)
```
